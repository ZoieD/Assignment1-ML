#### Correlation analysis for numerical columns

```python
corr = data.corrwith(data['SalePrice'])
# print(corr)
# Filter rows where the correlation is greater than certain percentage
filtered_corr = corr[corr > 0.6]

# Print the result
print(filtered_corr)

# Count the number of rows with corr > 0.2
count_filtered = filtered_corr.count()

# Total number of rows in the correlation series
total_rows = corr.count()

# Calculate proportion
proportion = count_filtered / total_rows

# Print the results
print(f'Rows with corr > 0.2: {count_filtered} out of {total_rows}')
print(f'Proportion: {proportion:.2%}')
```

#### Data cleaning for selected features

```python
seleted_features = ['Overall Qual', 'Total Bsmt SF','1st Flr SF','Gr Liv Area', 
                   'Garage Cars','Garage Area','SalePrice']
```

```python
data_selected = data[seleted_features]
data_selected
```

##### 1.Check on the nulls

```python
# Check for null values in the selected features
null_values = data_selected.isnull().sum()

# Calculate the percentage of missing values for each column
null_percentage = (null_values / len(data_selected)) * 100

# Combine the counts and percentages into a dataframe for analysis
null_analysis = pd.DataFrame({
    'Missing Count': null_values,
    'Missing Percentage': null_percentage,
    'Total Rows': len(data_selected)
})

# Display the null analysis
print(null_analysis)
```

##### From the nulls analysis, we can see that there are only one row missing in the columns, so we decided to just drop the nulls.

```python
data_cleaned = data_selected.dropna()
data_cleaned
```

#### Export to csv for model tranining
